```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
R = 1
n = 150

x = np.linspace(-R, R, n + 1)
y = np.linspace(-R, R, n + 1)
```

```python
def esfera(x, y):
    R = 1
    z = R**2 - x**2 - y**2
    if z <= 0.:
        return 0.
    else:
        return np.sqrt(z)
```

```python
esfera = np.vectorize(esfera)
X, Y = np.meshgrid(x, y)
Z = esfera(X, Y)
```

```python
def volumen(Z):
    vol = 0
    n = len(Z) - 1
    for i in range(n):
        for j in range(n):
            v1 = Z[i][j]
            v2 = Z[i][j + 1]
            v3 = Z[i + 1][j]
            v4 = Z[i + 1][j + 1]
            P = (v1 + v2 + v3 + v4) / 4
            vol += P
    return vol
```

```python
print(round(volumen(Z), 2))
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```
