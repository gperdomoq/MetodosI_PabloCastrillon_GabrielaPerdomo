```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
a = 0
b = 1000
n = 11
```

```python
def black_body_radiation(x):
    return (x**3)*np.exp(x) / (np.exp(x) - 1.0)
```

```python
def gauss_laguerre(f, a, b, n):
    x, w = np.polynomial.laguerre.laggauss(n)
    z = 0
    for i in range(n):
        t = (0.5) * (2) * (x[i])
        z += w[i] * f(t)
    return z
```

```python
Int = gauss_laguerre(black_body_radiation, a, b, 3)
print("Integral: ", Int)
```

```python
def error_points(n):
    error = np.array([])
    print(n)
    for i in n:
        error = np.append(error, gauss_laguerre(black_body_radiation, a, b, int(i)))
    return error
```

```python
n = np.linspace(2, 10, 9)
error = error_points(n)

plt.plot(n, error, 'o')
plt.grid()
plt.xlabel('n')
plt.ylabel('error')
plt.title('Error vs n')
plt.show()
```
